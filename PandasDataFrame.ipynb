```python
# Multi indeks
ilkIndex = ["Türkiye", "Türkiye", "Türkiye", "ABD", "ABD", "ABD"]
icIndex  = ["İstanbul", "Ankara", "Sivas", "New York", "Washington DC", "Teksas"]
```

```python
birlesmisIndex = list(zip(ilkIndex,icIndex))
```

```python
birlesmisIndex
```

```python
birlesmisIndex = pd.MultiIndex.from_tuples(birlesmisIndex)
```

```python
birlesmisIndex            # MultiIndex olarak tanımla..
```

```python
type(birlesmisIndex)
```

```python
numaralar = [[34, "otuzdört"], [ 6, "altı"], [58, "ellisekiz"], [32,"otuziki"], [5,"beş"],[ 57, "elliyedi"]]
numaralarNumpyDizisi = np.array(numaralar)
```
