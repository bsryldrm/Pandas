# Merge / Birleştirmek

```python
mergesozluk = {"Isim": ["Ahmet", "Mehmet", "Zeynep","Atıl"],
               "Spor": ["Koşu", "Yüzme", "Koşu", "Basketbol"]}
```

```python
mergeDataFrame1 = pd.DataFrame(mergesozluk) 
```

```python
mergeDataFrame1
```

```python
mergesozluk2 = {"Isim": ["Ahmet", "Mehmet", "Zeynep","Atıl"],
               "Kalori": [100,200,150,250]}
```

```python
mergeDataFrame2 = pd.DataFrame(mergesozluk2) 
```

```python
mergeDataFrame2
```

```python
pd.merge(mergeDataFrame1, mergeDataFrame2, on ="Isim")     #hangi kolon üzerinden işlem yapmak istediğimizi belirtiriz.
```
