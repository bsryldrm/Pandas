# Series/Seriler

Pandas, veriyi topladıktan sonra veriyi analize uygun hale getirmek, veri analizi için oldukça önemlidir.
Pandas kütüphanesi ile veri organize edilerek analizler daha kolay ve hızlı yapılır.
Pandas Python’ın en önemli kütüphanelerinden biridir. kaynakça:https://tirendazakademi.medium.com/pandas-kütüphanesi-nedir-905012e654b8

```python
istanbul = {"bir":1, "dört":4, "Beş":5, "üç":3 }
```

```python
pd.Series(istanbul)
```

```python
yazı =["bir","dört","beş","üç"]
sayı =[1,4,5,3]
```

```python
pd.Series(yazı,sayı)
```

```python
numpyDizisi = np.array([1,4,5,3])
```

```python
numpyDizisi
```

```python
pd.Series(numpyDizisi)
```

```python
pd.Series(numpyDizisi, yazı)
```
