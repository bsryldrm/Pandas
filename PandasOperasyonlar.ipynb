# groupby
bizim sınıflarımızı gruplar. multiIndex olmasına gerek yok herhangi kolondan oluşturabiliriz.

```python
isYeri = {"Departman": ["Yazılım","Yazılım","Pazarlama","Pazarlama","Hukuk","Hukuk"],
         "Calisan Kişiler": ["Fatih", "Sultan", "Mehmet", "Büşra", "İlayda", "Zeynep"],
         "maaş": [100,200,300,400,500,600]}
```

```python
isYeriDataFrame = pd.DataFrame(isYeri)
```

```python
isYeriDataFrame
```

```python
grupObjesi = isYeriDataFrame.groupby("Departman")
```

```python
grupObjesi.count()        # saymak demektir. her departmanda ki kişi sayısı ve maaş alan sayısı verir.
```

```python
grupObjesi.mean()         #ortalama hesaplanır
```

```python
grupObjesi.min()
```

```python
grupObjesi.max()
```

```python
grupObjesi.describe()
```
